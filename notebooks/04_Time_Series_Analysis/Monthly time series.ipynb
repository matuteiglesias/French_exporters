```python
fig, axs = plt.subplots(3, 1, figsize = (12, 15))
for i, _file in enumerate([ 'ID_YM_nuq', 'CN_ID_8_YM_nuq']): #VAT_YM_nuq # '_YM',
    ax = axs[i]
    monthly_summary = pd.read_csv('./../../data/processed/'+_file+'.csv')

    monthly_summary.set_index(['IMPORT', 'YEAR', 'MONTH']).unstack([-2, -1]).T.plot(ax = ax)
    ax.grid(linestyle='--', color='gray', lw = .5)

    ax.set_xticks(arange(0, 12*17+1, 12))
    ax.set_xticklabels(range(1997, 2014))

# plt.savefig('./Figures/mn_total_trade.png')
```
