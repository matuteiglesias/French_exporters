## Growth distribution

```python
fig, axes = plt.subplots(2, 2, figsize = (15, 10))

    
for i in [0, 1]:
    ax = axes[i]

    firm_y = firm_mx_y.loc[firm_mx_y.IMPORT == i]

    firm_y_uns = firm_y.set_index(['ID', 'YEAR'])[['VART']].unstack()
    
    df = firm_y.set_index(['ID', 'YEAR'])[['log_VART']]

    df2 = pd.DataFrame(-log10(firm_y_uns).diff(axis = 1).stack())
    growth = pd.concat([df, df2], axis = 1).dropna()
    growth.columns = ['log_VART', 'growth']
    
    growth['size_bin'] = pd.cut(growth['log_VART'], arange(3, 9, .5))        

#     Growth scatter
    sample = growth.groupby('size_bin').apply(lambda x: x.sample(2000, replace = True))
    sample.plot(x = 'log_VART', y = 'growth', marker = 'o', mec = 'None', linewidth = 0, alpha = .05, ax = ax[0])
    ax[0].set_xlim(3, 8)
#     ax[0].legend_.remove()
    ax[0].set_xlabel('log firm size')
    ax[0].set_ylabel('log growth')
    ax[0].legend(fontsize = 12)
    if i == 0: ax[0].set_title('direct scatterplot')
        
    #Growth profiles
    growth['growth_bin'] = pd.cut(growth.growth, arange(-2, 3, .1))
    hist_data = growth.groupby(['size_bin', 'growth_bin']).count()[['log_VART']].rename(columns = {'log_VART': 'counts'}).dropna().reset_index()
    hist_data['bin_mid'] = [b.mid for b in hist_data['growth_bin'].values]
    
    for s in hist_data.size_bin.unique()[::-3]:
        hist_data.loc[hist_data.size_bin == s].plot(x = 'bin_mid', y = 'counts', label = s, ax = ax[1],
                                                            marker = 'o', linewidth = 0.1, alpha = .5)
    
    ax[1].set_yscale('log')
    ax[1].set_xlim(-2, 2)
    ax[1].set_xlabel('log growth')
    ax[1].set_ylabel('PDF')
    ax[1].legend(fontsize = 12)
    if i == 0: 
        ax[1].set_title('cuts by firm size')

plt.savefig('./../Figures/growth_example.png')
```

### Top 1% takes...

```python
from scipy.special import erfinv, erf

def erfinv_(x):
    return erfinv(2*(x - .5))

def erf_(x):
    return .5*erf(x)+.5

from numpy import log
```

```python
### CDF of counts and value

X = linspace(-3, 3, 100, endpoint=True)

plt.plot(X, erf_(X))
plt.plot(X, erf_(X - .5*log(10)))
```

```python
# Top x% of agents is taking y% of the value.

plt.figure(figsize = (8, 5))
plt.plot(erf_(X - .5*log(10)), erf_(X))
plt.xscale('log')
plt.yscale('log')
plt.xlim(1e-10, 1)
plt.title('log-normal implication \n for the distribution among the top')
plt.xlabel("Top 'x' fraction of agents...")
plt.ylabel("... take 'y' of the value")
matplotlib.rcParams.update({'font.size': 14})
# plt.savefig('./../Figures/top_lognormal_demo.png')
plt.show()
```
